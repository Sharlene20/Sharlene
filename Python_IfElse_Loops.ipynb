**Python If ...Else**

If

```python
a = 20
b = 48
if b > a:
  print("b is greater than a")
```

```python
a = 99
b = 990
if b > a:
  print("b is greater than a")
```

Elif

```python
a = 69
b = 69
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

```python
a = 899
b = 79
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

Else

```python
a = 599
b = 59
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

Short Hand If

```python
if a > b: print("a is greater than b")
```

Short Hand If..Else

```python
a = 9
b = 99
print("A") if a > b else print("B")
```

Leap Year

```python
year= eval(input("Please input the year"))

if((year%400==0) or (year%4==0 and year%100!=0)):
  print("{0} is a leap year" .format (year))
else: 
    print("{0} is not a leap year" .format(year))
```

**Loops**

```python
fruits = ["orange", "rambutan", "grapes"]
for x in fruits:
  print(x)
```

Looping Through a String

```python
for x in "rambutan":
  print (x)
```

**Range () Function**

```python
for x in range(5):
  print(x)
```

```python
for x in range(50):
  print(x)
  if x == 9:
    break
```

**While Loops**

```python
i = 3
while i < 9:
  print(i)
  i += 3
```

```python
i = 3
while i < 19:
  print(i)
  if i == 7:
    continue
  i += 3
```

```python
i = 3
while i < 19:
  print(i)
  if i == 7:
    break
  i += 3
```

For Loops

```python
for x in range(99):
  if x == 9:
    continue

  print(x)
```

```python
for x in range(99):
  if x%9 == 19:
    continue

  print(x)
```

Infinite Loop

```python
i = 3
while i < 19:
  print(i)
  if i == 6:
    continue
   # break
  i += 3
```
